```python
df=pd.read_csv("Zingaro_Company.csv")
df.head()
```

```python
df.shape
```

```python
df.describe().T
```

```python
from scipy.stats import ttest_1samp
t_statistic,p_value=stats.ttest_1samp(df['Unpolished '],150,alternative='less')
print('t_test',t_statistic)
print('p-value',p_value)
```

```python

alpha_value=0.05
if p_value<alpha_value:
    print("reject null hypothesis")
else:
    print("fail to reject null hypothesis")
        
```

```python
import math
from scipy import stats
from scipy.stats import ttest_ind
t_stat,p_val= ttest_ind(df["Unpolished "],df["Treated and Polished"])
print(t_stat)
print(p_val)
```

```python

alpha_value=0.05
if p_value<alpha_value:
    print("reject null hypothesis")
else:
    print("fail to reject null hypothesis")
```

```python
df1.Dentist=pd.Categorical(df1.Dentist)
df1.Method=pd.Categorical(df1.Method)
df1.Alloy=pd.Categorical(df1.Alloy)
```

```python
alloy1=df1[df1['Alloy']==1]
alloy2=df1[df1['Alloy']==2]
```

```python
from scipy.stats import shapiro
name=alloy1.get('Dentist')
for name in alloy1['Dentist'].unique():
    print("{}: {}".format(name,shapiro(alloy1['Response']
                                       [alloy1['Dentist']==name])))
```

```python
name=alloy2.get('Dentist')
for name in alloy2['Dentist'].unique():
    print("{}: {}".format(name,shapiro(alloy2['Response']
                                       [alloy2['Dentist']==name])))
```

```python
formula='Response ~ C(Dentist)'
model=ols(formula,alloy1).fit()
aov_table=anova_lm(model)
print(aov_table)
```

```python
formula='Response ~ C(Dentist)'
model=ols(formula,alloy2).fit()
aov_table=anova_lm(model)
print(aov_table)
```

```python
mc=MultiComparison(df1['Response'],df1['Dentist'] )
result=mc.tukeyhsd()
print(result)
```
